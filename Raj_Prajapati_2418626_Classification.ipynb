```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
df.columns = df.columns.str.strip()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['is_safe'], bins=30, kde=True)
plt.title('Water Safety Analysis')
plt.xlabel('Water Safety (0 = Unsafe, 1 = Safe)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Summary statistics
summary_statistics = df.describe()
```

```python
# Visualizing distributions of numerical variables
plt.figure(figsize=(14, 12))
df.hist(bins=30, figsize=(14, 12))
plt.tight_layout()
plt.show()
```

```python
# Identifying outliers using boxplots
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, palette="Set2")
plt.xticks(rotation=45)
plt.title("Boxplot of Features to Identify Outliers")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```
